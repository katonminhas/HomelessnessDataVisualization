```python
import pandas as pd
```

```python
path = "C:/Users/Katon/Documents/JHU/DataVisualization/final_project/data/final_data/"

# non-projections
population = pd.read_excel(path+"population.xlsx")#.set_index('State')
housing = pd.read_excel(path+"housing.xlsx")#.set_index('State')
homeless = pd.read_excel(path+"homeless.xlsx")#.set_index('State')
hpi = pd.read_excel(path+"hpi.xlsx")#.set_index('State')
```

```python
population.head(3)
```

```python
population_long = pd.melt(population, id_vars='State', var_name='Year', value_name='Population').set_index(['State', 'Year'])
housing_long = pd.melt(housing, id_vars='State', var_name='Year', value_name='Housing').set_index(['State', 'Year'])
homeless_long = pd.melt(homeless, id_vars='State', var_name='Year', value_name='Homeless').set_index(['State', 'Year'])
hpi_long = pd.melt(hpi, id_vars='State', var_name='Year', value_name='HPI').set_index(['State', 'Year'])
```

```python
features_df = pd.concat([population_long, housing_long, homeless_long, hpi_long], axis=1)
```

```python
features_df.isnull().sum(axis=1).astype(bool).sum()
```

```python
features_df.to_excel(path+'all_features.xlsx')
```

```python
# projections
hpi_projections = pd.read_excel(path+'final_hpi_projections.xlsx', header=[0,1])
homeless_projections = pd.read_excel(path+'final_homeless_projections.xlsx', header=[0,1])
```

```python
hpi_projections.set_index(('Unnamed: 0_level_0', 'Percent of Current Housing Construction'), inplace=True)
homeless_projections.set_index(('Unnamed: 0_level_0', 'Percent of Current Housing Construction'), inplace=True)
```

```python
hpi_projections.head()
```

```python
hpi_projections.loc[25, ('NE', 2023)]
```

```python
unique_states = list(set([col_name[0] for col_name in hpi_projections.columns]))
unique_years = list(set([col_name[1] for col_name in hpi_projections.columns]))
```

```python
state_list, year_list, pct_list, hpi_list, homeless_list = [], [], [], [], []
for state in unique_states:
    for year in unique_years:
        for pct in hpi_projections.index:
            state_list.append(state)
            year_list.append(year)
            pct_list.append(pct)
            hpi_list.append(hpi_projections.loc[pct, (state, year)])
            homeless_list.append(homeless_projections.loc[pct, (state, year)])
```

```python
projections_df_final = pd.DataFrame([state_list, year_list, pct_list, hpi_list, homeless_list]).T
projections_df_final.columns = ['State', 'Year', 'Percent_of_Current_Housing_Construction', 'HPI', 'Homeless']
```

```python
projections_df_final.to_excel(path+'all_projections.xlsx')
```
